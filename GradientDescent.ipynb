# Gradient Descent for Linear Regression

Gradient descent is a function that will help us "move" and "rotate" our 2D line to fit the data as well as possible. The gradient descent algorithm needs a cost function in order to keep moving the line until we find the best w and b parameters.

Recall that the cost function for linear regression is given as: $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2$$ 

And the gradient descent algorithm itself is described as:
$$\begin{align*} \text{repeat}&\text{ until convergence:} \; \lbrace \newline
\;  w &= w -  \alpha \frac{\partial J(w,b)}{\partial w}\; \newline 
 b &= b -  \alpha \frac{\partial J(w,b)}{\partial b}  \newline \rbrace
\end{align*}$$

This means that for in order to use this gradient descent algorithm, we need to find the partial derivatives of the cost function in respect to w and b.

Here I have skipped the difficult math for you and this is the results of those derivatives:

$$
\begin{align}
\frac{\partial J(w,b)}{\partial w}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})x^{(i)}\\
  \frac{\partial J(w,b)}{\partial b}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})\\
\end{align}
$$

Given these formulas, we can now implement the two functions that will help us get our gradient descent computation.

```python
# We start by making a function to compute the partial derivatives. This can be done like this:

def compute_gradients(x, y, w, b):

    m = x.shape[0] #recall you can also use len()
    dj_dw = 0
    dj_db = 0

    for i in range(m):
        #Recall that f_wb(x^i) = wx+b
        f_wb = w * x[i] + b
        dj_dw += (f_wb - y[i]) * x[i]
        dj_db += (f_wb - y[i])

    dj_dw = (1 / m) * dj_dw
    dj_db = (1 / m) * dj_db

    return dj_dw, dj_db
```

```python
# Now we can use that function to compute the whole gradient descent algorithm, using that function inside of this one:

def gradient_descent(x, y, w_in, b_in, alpha, iters):

    w = w_in
    b = b_in

    for i in range(iters):

        dj_dw, dj_db = compute_gradients(x, y, w, b) #Important to define both exactly how the compute_gradient function returns them!

        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        if i % math.ceil(iters / 10) == 0:
            print(f"Iteration: {i}: Cost: {compute_cost(x, y, w, b)} ")

    return w, b
    
```
