# Cost Function

The cost function is a useful function that will help us find the optimal values for w and b. The cost function itself measures how well our model is predicting the target.

The basic equation for the cost with only one variable looks as follows: $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}
 ) - y^{(i)})^2$$ 

 where 
  $$f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{2}$$

  So how can we actually implement this idea in code? Let's take a look:


```python
#Import libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make cost compute function:

x = np.array([2, 5, 4, 6, 7])
y = np.array([5, 8, 7, 7, 10])
```

```python
def compute_cost(x, y, w, b):

    m = x.shape[0] #This only works if x is a np array. if its a list, we will be better using len().

    total_sum = 0

    for i in range(m):

        f_wb = w * x[i] + b
        total_sum += (f_wb - y[i])**2

    return (1 / (2 * m)) * total_sum
```

This is how we would succesfully implement a cost function for our data!
